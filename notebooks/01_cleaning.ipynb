```python
import pandas as pd
import numpy as np

# Change the filename below to match YOUR actual CSV file name
df = pd.read_csv('../data/raw/train.csv', encoding='latin1')
print(df.shape)
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Fix date columns
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

# Add new useful columns (only using Sales since no Profit/Discount)
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.strftime('%B')
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

print("Cleaned! Shape:", df.shape)
print(df.head())
```

```python
df.to_csv('../data/cleaned/superstore_cleaned.csv', index=False)
print("✅ Saved to cleaned folder!")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/Sample - Superstore.csv', encoding='latin1')
print(df.shape)
df.head()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Fix date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Add new useful columns
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.strftime('%B')
df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

print("Cleaned! Shape:", df.shape)
df.head()
```

```python
df.to_csv('../data/cleaned/superstore_cleaned.csv', index=False)
print("✅ Saved to cleaned folder!")
```
